```python
# 11672
# file name match
from fnmatch import fnmatch

def count_evens(number:int)->int:
    count = 0
    for digit in str(number):
        if int(digit) % 2 == 0:
            count += 1
    return count 

# n - всего кол-во
# evens - кол-во четных
# n - evens - кол-во нечетных
# even == n - evens | + evens
# 2*evens == n

# 12*34?5
for number in range(0,10**10,21025):
    st = str(number)
    evens = count_evens(number)
    odds = len(st) - evens
    if fnmatch(st,"12*34?5") and evens == odds:
        print(number, number / 21025)
```
