```python
# 12253 
def f(start,stop, stops:list):
    if start > stop or start in stops:
        return 0
    if start == stop:
        return 1
    return f(start+1,stop,stops) + f(start+3,stop, stops) + f(start**2,stop, stops)

print(f(3,20,[16]))
print(f(20,26,[]))
print(156*6)
```
