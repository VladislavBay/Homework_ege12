```python
# https://inf-ege.sdamgia.ru/problem?id=60247
def f(n):
    n_bin = bin(n)[2:]
    if n % 3 == 0:
        n_bin += n_bin[-3:]
    else:
        n2 = bin((n%3)*3)[2:]
        n_bin += n2
    return int(n_bin,2)  

# print(f(12)) # 100
# print(f(4)) # 19
lst = []
for i in range(1,1000):
    r = f(i)
    if r > 151:
        lst.append(r)

print(min(lst))
    
```

```python
# https://inf-ege.sdamgia.ru/problem?id=56505
def f(n):
    n_bin = bin(n)[2:]
    st = str(n)
    if sum(map(int,st)) % 2 == 0:
        n_bin += "0"
    else:
        n_bin +="1"
    return int(n_bin,2)


print(f(f(f(17))))
# [123_456_789; 1_987_654_321]  

number = 100_000
r = f(f(f(number)))
while r < 123_456_789:
    number += 100_000
print(number)

```
